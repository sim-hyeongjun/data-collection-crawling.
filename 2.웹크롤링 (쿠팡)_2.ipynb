#### 다중 페이지 한번에 크롤링 하는법

```python

for page_num in range(10):
    if page_num== 0:
        res=requests.get('https://davelee-fun.github.io/')
    else:
        res = requests.get('https://davelee-fun.github.io/page'+str(page_num+1))
    soup=BeautifulSoup(res.content,'html.parser')

    items=soup.select('h4.card-text')
    for item in items:
        print(item.get_text())
```

### 네이버 증시 크롤링

#### 상위 탑종목 크롤링

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://finance.naver.com/sise')
html = res.content.decode('euc-kr', 'replace')
soup = BeautifulSoup(html, 'html.parser')
items = soup.select('#siselist_tab_0>tr>td>a')
for item in items:
    print(item.get_text())
#siselist_tab_0
```

##### 갭상승 상위 크롤링

```python
res = requests.get('https://finance.naver.com/sise/')

html = res.content.decode('euc-kr', 'replace')
soup = BeautifulSoup(html, 'html.parser')

#갭상승 종목명 순위
data = soup.select('#trend_tab_0 >tr> td > a')
#갭상승 종목 상승폭 순위
data1 = soup.select('#trend_tab_0 > tr > td.number')

for item in data:
    print(item.get_text())

for item1 in data1:
    print(item1.get_text())

#trend_tab_0 > tbody > tr> td > a
#trend_tab_0 > tbody > tr:nth-child(1) > td.number
```
