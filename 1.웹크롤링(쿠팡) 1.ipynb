## 가져온 데이터 원하는 형태로 전처리 과정

```python
##가장 쉬운 사이트 데이터 크롤링
import requests
res=requests.get('http://davelee-fun.github.io/blog/crawl_test')
soup=BeautifulSoup(res.content, 'html.parser')

 # 가장 중요 'ul',id='dev_course_list'
sel= soup.find('ul',id='dev_course_list')
data= sel.find_all('li',class_="course")


# 문장중에 내가 원하는 문장만 뺴서 출력하는 방법
#  split()으로 - 기준으로 리스트로 나누고 인데스 -1 자리만 슬라이싱
#  split()으로 [기준으로 문장과 [숫자]를 리스트로 나눈후 인덱스0자리만 슬라이싱
#  strip()으로 슬라이싱한 문장에 불필요하는 공백 제거
for index, se in enumerate(data):
    print(str(index+1)+ '.',se.get_text().split('-')[-1].split('[')[0].strip())
    print(str(index+1)+ '.',se.get_text().split('-')[-1].split('[')[0])
```

#### 중급

```python
from bs4 import BeautifulSoup
import requests

res = requests.get('https://davelee-fun.github.io/blog/crawl_html_css.html')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('tr')
for item in items:
    columns=item.select('td')
    row_str =''
    for column in columns:
       row_str +=  column.get_text()
    print (row_str)
```

#### 고급

```python
from bs4 import BeautifulSoup
import requests
res = requests.get('https://davelee-fun.github.io/blog/crawl_html_css.html')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('tr')
for item in items:
   columns = item.select('td')
   row_str = ''
   for column in columns:
     row_str += ', ' + column.get_text()
   print (row_str)
```

#### 고급에서불필요한 문자 제거 (,)

```python
from bs4 import BeautifulSoup
import requests
res = requests.get('https://davelee-fun.github.io/blog/crawl_html_css.html')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('tr')
for item in items:
   columns = item.select('td')
   row_str = ''
   for column in columns:
     row_str += ', ' + column.get_text()
   print (row_str[2:])
```
